```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder # Convert Categorical Features to Numerical
from sklearn.model_selection import train_test_split # Split the data into testing and training
import keras
from keras.models import Sequential
from keras.layers import Dense # Convert input from each layer
from keras.layers import Dropout
from keras.layers import LeakyReLU,PReLU,ELU # Activation Functions
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Unique columns-drop
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
# Convert Categorical Data to Numerical
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Geography']=le.fit_transform(df['Geography'])
```

```python
df.head()
```

```python
# Feature Extractions
y=df.pop('Exited')
x=df
```

```python
# Split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Initialize the ANN
classifier = Sequential()
```

```python
# Add the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu',input_dim = 10))
```

```python
# Add the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
# Add the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
# Compile the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
model_history=classifier.fit(x_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Evaluate the model
classifier.summary()
```
